## Diffusion Modeling 

using the diffusion equation to measure how many 
cosmic rays will make its way to earth from SN at different distances



n(E, r ,t )=  Q(E) / ((np.pi**(3/2)) * (rdiff**3) * exp((-r**2)/(rdiff**2)) 


```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

```python
def diffusion(E,r,t, **kwargs):
    Q  = 10e50 * ((E**-2.2)* np.exp((-E)/1e6))
    D  = (7*10e-2) * (E/(10))**(1/3)
    rd = (4 * D * t) **.5
    n  = (Q/ (((np.pi**(3/2)) * (rd**3)))) * (np.exp(-(r**2) / rd**2))
    return n

r1 = np.arange(.1,13, .01) #pc

```
